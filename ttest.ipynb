```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
sns.set(style='whitegrid', context='talk', palette='rainbow')
plt.rcParams['grid.color'] = 'gray'
plt.rcParams['grid.linestyle'] = '--'
plt.rcParams['grid.linewidth'] = 0.5
plt.style.use('dark_background')
```

```python
clivestock = pd.read_csv("./data/Azerbaijan_crops_and_livestocs_products.csv")
valueof_prod = pd.read_csv("./data/Azerbaijan_Value_of_Agricultural_Production.csv")
```

```python
display(clivestock.head())
display(valueof_prod.head())
```

```python
display(clivestock.columns)
display(valueof_prod.columns)
```

```python
clivestock = clivestock[["Element","Item","Year","Value","Unit","Flag Description"]]
valueof_prod = valueof_prod[["Element","Item","Year","Value","Unit","Flag Description"]]
```

```python
clivestock_production = clivestock[clivestock["Element"] =="Production"]
clivestock_production = clivestock_production[~(clivestock_production["Value"].apply(lambda x : x==0))]
clivestock_production = clivestock_production[["Year","Item","Value"]]
clivestock_production
```

```python
production_17_22 = clivestock_production.query("Year >= 2017 and Year<=2022")
production_92_96 = clivestock_production.query("Year >= 1992 and Year<=1996")

t_stat, p_value = ttest_ind(production_92_96[["Year","Value"]], production_17_22[["Year","Value"]])
print(t_stat,p_value)
```

P value 5 dən kiçik olduğu üçün statistiki olaraq test mənalıdı.1992-1996 cı illərlə 2017-2022 illəri arasında statistiki məna əlaqəsinin olduğunu deyə bilərik.Yəni bu illər üzrə verilən dəyərlər(Values) təsadüfiliyi çox az ehtimal edilir.
